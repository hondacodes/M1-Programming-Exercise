Description of the Programming Exercise:
    
    The dataset used for this programming exercise covers the Philippines' Dengue Cases from 2016 to 2020.  The dataset was taken from Kaggle and uploaded by the user, Jhon Vincent Gupo, in 2022 (as it was updated two years ago).  The dataset included all dengue cases covered in all regions in the Philippines; hence, for this programming exercise, the data was limited to only showing the dengue statistics in NCR.  

    In further organising the dataset, the student sorted the NCR dengue statistics per year; hence, the five tables with captions indicated the data in the respective tables.  The operations made to the dataset included the computation for each year's survival and death rate for dengue per month.  In addition, the averages for these two rates were taken and displayed in the table.

    In analysing each table, 2017 to 2019 saw a stable behaviour for the survival and death rate for Dengue with the survival rate being at around 99.4% to 99.5%.  However, 2016 and 2020 saw a slump in survival rate as their survival rates were 78.03% and 76.59%, respectively.  The slump in survival rate was attributed to the rise in cases and deaths around the fourth quarter of each year (September to December).  It is to be noted that the fourth quarters of 2017 to 2019 did not show any spikes in Dengue cases and deaths, this can be interpretted that the handle in Dengue cases was better during this period as compared to 2016 and 2020.  2020 was a particular year as brought on by the COVID-19 pandemic where lockdowns were in place which could explain the 100% survival rate from April 2020 to July 2020, but it could also be due to other factors which can include cases not being recorded and the like.
    


```python
import requests as rq
from io import BytesIO
import numpy as np
import pandas as pd
```

```python
dataset = "https://raw.githubusercontent.com/hondacodes/M1-Programming-Exercise/main/ph_dengue_cases2016-2020.csv"
dsg = rq.get(dataset).content
dsr = pd.read_csv(BytesIO(dsg))
pd.set_option('display.max_row',None)
dsr.head()
```

```python
#Sorting NCR data per year (2016), computing survival and death rate, computing average survival and death rate, and cleaning table
NCR2016 = dsr.loc[dsr['Region'].str.contains('NCR')]
NCR2016 = NCR2016.loc[NCR2016.Year == 2016,"Month":"Dengue_Deaths"]
NCR2016['Dengue Survival Rate'] = np.round((NCR2016['Dengue_Cases']-NCR2016['Dengue_Deaths'])/NCR2016['Dengue_Cases']*100,2)
NCR2016['Dengue Death Rate'] = np.round((NCR2016['Dengue_Deaths'])/NCR2016['Dengue_Cases']*100,2)
NCR2016 = NCR2016.reset_index()
NCR2016 = NCR2016.drop(columns=['index','Region','Dengue_Cases','Dengue_Deaths'])
print('\033[1mNCR Dengue Statistics for 2016\033[0m')
print('')
print('\033[1mAverage Survival Rate: \033[0m',np.round(NCR2016['Dengue Survival Rate'].mean(),2),'%')
print('\033[1mAverage Death Rate: \033[0m',np.round(NCR2016['Dengue Death Rate'].mean(),2),'%')
NCR2016
```
